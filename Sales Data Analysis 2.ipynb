# What US city had The Highest Number Of Sales ?

# First,Adding a City Column

```python
dataframe.head()
```

```python
#.apply() function in pandas is used to pass a function and apply it to every value.
#It helps to separate data based on the conditions required.
```

```python
#We use lambda x because it captures all the cell contents

dataframe['City']=dataframe['Purchase Address'].apply(lambda x: x.split(',')[1])
                                                                        
```

```python
#We define states to avoid instances of duplicates

def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
```

```python
dataframe['City']=dataframe['Purchase Address'].apply(lambda x: f"{get_city(x)} {get_state(x)}")
```

```python
dataframe.head()
```

```python
results_city=dataframe.groupby('City').sum()
results_city
```

```python
import matplotlib.pyplot as plt

cities=[city for city, df in dataframe.groupby('City')]

plt.bar(cities,results_city['Sales'])
plt.xticks(cities,rotation='vertical',fontsize=12)
plt.ylabel('Sales in US Dollars($)')
plt.xlabel('City name')
plt.show()
```
